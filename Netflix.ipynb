```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("/content/netflix_tv_shows_detailed_up_to_2025.csv")
print(data.describe())
```

```python
print(data.columns)
```

```python
movies = len(data[data["type"] == "Movie"])
tvshows = len(data[data["type"] == "TV Show"])

# Step 4: Store data for plotting
labels = ["Movies", "TV Shows"]
counts = [movies, tvshows]

# Step 5: Create bar chart
plt.bar(labels, counts)
plt.bar(labels, counts)

# Step 6: Add labels and title
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```
